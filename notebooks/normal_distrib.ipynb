```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/noahgift/regression-concepts/master/height-weight-25k.csv')
df.describe()
```

```python
import seaborn as sns
import numpy as np
HEIGHT_INCHES = "Height-Inches"
WEIGHT_POUNDS = "Weight-Pounds"
```

```python
sns.lmplot(HEIGHT_INCHES, WEIGHT_POUNDS, data=df)
```

```python
sns.jointplot(HEIGHT_INCHES, WEIGHT_POUNDS, data=df, kind="kde")
```
